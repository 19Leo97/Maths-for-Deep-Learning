##**Problem 7.8**
###**Answer**
Using ReLU as an activation function, we can obtain points in the loss function where derivatives for the activation function with respect to its respective preactivation function can be 1 or 0. By using the gradient, we can determine the direction in which the loss function value decreases. However, if we implement the rectangular or heaviside functions, the derivatives of the activation function with respect to its respective preactivation function are always 0. Consequently the gradient also will always be 0, failing in providing us with information about the direction in which the loss function value decreases. It is problematic since we cannot use analytic methods to find the gradient, forcing us to use other kind of methods such as finit differences to compute it.

##**Problem 7.13**
###**Answer**

\begin{align}
\frac{\partial f_1}{\partial x} &= \exp(x)\\
\frac{\partial f_2}{\partial x} &= \frac{\partial f_2}{\partial f_1} \cdot \frac{\partial f1}{\partial x} = 2f_1 \exp(x)\\
\frac{\partial f_3}{\partial x} &= \frac{\partial f_3}{\partial f_2} \cdot \frac{\partial f2}{\partial x} + \frac{\partial f_3}{\partial f_1} \cdot \frac{\partial f1}{\partial x} = \frac{\partial f2}{\partial x} + \frac{\partial f1}{\partial x} = \exp(x) (2f_1 + 1)\\
\frac{\partial f_4}{\partial x} &= \frac{\partial f_4}{\partial f_3} \cdot \frac{\partial f3}{\partial x} = \exp(f_3)\exp(x) (2f_1 + 1)\\
\frac{\partial f_5}{\partial x} &= \frac{\partial f_5}{\partial f_3} \cdot \frac{\partial f3}{\partial x} = \cos(f_3)\exp(x) (2f_1 + 1)\\
\frac{\partial y}{\partial x} &= \frac{\partial y}{\partial f_5} \cdot \frac{\partial f5}{\partial x} + \frac{\partial y}{\partial f_4} \cdot \frac{\partial f4}{\partial x} = \frac{\partial f5}{\partial x} + \frac{\partial f4}{\partial x} = \exp(x) (2f_1 + 1)[\cos(x)+\exp(f_3)]\\
\end{align}

We don't use forward-mode differentiation when computing the gradients of deep neural networks because we not only compute the gradient of $y$ with respect to $x$, but also need to calculate the gradient of $y$ with respect to the parameters $\mathbf{θ}$ and $\mathbf{β}$. In this case, it is easier to compute $\frac{\partial y}{\partial \theta_i}$ and $\frac{\partial y}{\partial \beta_i}$, from the partial derivative $\frac{\partial y}{\partial f_i}$, than to have only calculated $\frac{\partial f_i}{\partial x}$, because in the latter case, this calculation isn't useful.

##**Problem 7.14**
###**Answer**

\begin{align}
E[b^2] &= \int_{-∞}^{∞} \text{ReLU}[a]^2 Pr(a) da\\
&= \int_{-∞}^{0} \text{ReLU}[a]^2 Pr(a) da + \int_{0}^{∞} \text{ReLU}[a]^2 Pr(a) da\\
&= 0 + \int_{0}^{∞} \text{ReLU}[a]^2 Pr(a) da\\
&= \frac{E[a^2]}{2}= \frac{σ^2 + E[a]^2}{2} = \frac{σ^2 + 0}{2} = \frac{σ^2}{2}
\end{align}

We have used some identities, such as $σ^2 = E[a^2] - E[a]^2$ and:

\begin{align}
\int_{0}^{∞} a^2 Pr(a) da = \frac{E[a^2]}{2}
\end{align}

Let's proof the latter:
\begin{align}
E[a^2] &= \int_{-∞}^{∞} a^2 Pr(a) da = \int_{-∞}^{0} a^2 Pr(a) da + \int_{0}^{∞} a^2 Pr(a) da\\
&= \int_{0}^{∞} a^2 Pr(a) da + \int_{0}^{∞} a^2 Pr(a) da\\
\end{align}
that is due to $a^2Pr(a)$ being an even function, as $a$ is symmetric, implying $a^2Pr(a) = a^2Pr(-a)$. Then, we have:
\begin{align}
E[a^2] &= 2 \int_{0}^{∞} a^2 Pr(a) da\\
\frac{E[a^2]}{2} &= \int_{0}^{∞} a^2 Pr(a) da
\end{align}

##**Problem 7.16**
###**Answer**

```python
import matplotlib.pyplot as plt
import torch, torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from torch.optim.lr_scheduler import StepLR

losses = []

# define input size, hidden layer size, output size
D_i, D_k, D_o = 10, 40, 5
# create model with two hidden layers
model = nn.Sequential(
    nn.Linear(D_i, D_k),
    nn.ReLU(),
    nn.Linear(D_k, D_k),
    nn.ReLU(),
    nn.Linear(D_k, D_o))

#He initialization of weights
def weights_init(layer_in):
  if isinstance(layer_in, nn.Linear):
    nn.init.kaiming_normal_(layer_in.weight)
    layer_in.bias.data.fill_(0.0)
model.apply(weights_init)

# choose least squares loss function
criterion = nn.MSELoss()
# construct SGD optimizer and initialize learning rate and momentum
optimizer = torch.optim.SGD(model.parameters(), lr = 0.1, momentum=0.9)
#object that decreases learning rate by half every 10 epochs
scheduler = StepLR(optimizer, step_size=10, gamma=0.5)

# create 100 random data points and store in data loader class
x = torch.randn(100, D_i)
y = torch.randn(100, D_o)
data_loader = DataLoader(TensorDataset(x,y), batch_size=10, shuffle=True)

#loop over the dataset 100 times
for epoch in range(100):
  epoch_loss = 0.0
  # loop over batches
  for i, data in enumerate(data_loader):
    #retrieve inputs and labels for this batch
    x_batch, y_batch = data
    # zero the parameter gradients
    optimizer.zero_grad()
    # forward pass
    pred = model(x_batch)
    loss = criterion(pred, y_batch)
    # backward pass
    loss.backward()
    #SGD update
    optimizer.step()
    # update statistics
    epoch_loss += loss.item()
  # print error
  #print(f'Epoch {epoch:5d}, loss {epoch_loss:.3f}')
  losses.append(epoch_loss)
  # tell scheduler to consider updating learning rate
  scheduler.step()

#Plotting the losses
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epochs')
plt.show()
```

```python
import matplotlib.pyplot as plt
import torch, torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from torch.optim.lr_scheduler import StepLR

losses = []

# define input size, hidden layer size, output size
D_i, D_k, D_o = 10, 40, 1
# create model with two hidden layers
model = nn.Sequential(
    nn.Linear(D_i, D_k),
    nn.ReLU(),
    nn.Linear(D_k, D_k),
    nn.ReLU(),
    nn.Linear(D_k, D_o),
    nn.Sigmoid()) #Sigmoid function aplied

#He initialization of weights
def weights_init(layer_in):
  if isinstance(layer_in, nn.Linear):
    nn.init.kaiming_normal_(layer_in.weight)
    layer_in.bias.data.fill_(0.0)
model.apply(weights_init)

# choose binary cross entropy loss function
criterion = nn.BCELoss()
# construct SGD optimizer and initialize learning rate and momentum
optimizer = torch.optim.SGD(model.parameters(), lr = 0.1, momentum=0.9)
#object that decreases learning rate by half every 10 epochs
scheduler = StepLR(optimizer, step_size=10, gamma=0.5)

# create 100 random data points and store in data loader class
x = torch.randn(100, D_i)
y = torch.randint(0, 2, (100, D_o)).float() #binary data point
data_loader = DataLoader(TensorDataset(x,y), batch_size=10, shuffle=True)

#loop over the dataset 100 times
for epoch in range(100):
  epoch_loss = 0.0
  # loop over batches
  for i, data in enumerate(data_loader):
    #retrieve inputs and labels for this batch
    x_batch, y_batch = data
    # zero the parameter gradients
    optimizer.zero_grad()
    # forward pass
    pred = model(x_batch)
    loss = criterion(pred, y_batch)
    # backward pass
    loss.backward()
    #SGD update
    optimizer.step()
    # update statistics
    epoch_loss += loss.item()
  # print error
  #print(f'Epoch {epoch:5d}, loss {epoch_loss:.3f}')
  losses.append(epoch_loss)
  # tell scheduler to consider updating learning rate
  scheduler.step()

#Plotting the losses
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epochs')
plt.show()
```
