##**Problem 8.6**
###**Answer**

We have used "Lecture Notes on High-Dimensional Data" by Sven-Ake Wegner as a reference for this proof. First of all, we need to state the following theorems:

If we have a random variable $X \sim \mathcal{N}(0,1,\mathbb{R}^d)$ where $\mathcal{N}(0,1,\mathbb{R}^d)$ is a standard normal multivariate distribution, where $0 = \mathrm{μ}$ and $1 = V[X] = σ^2$. $V$ stands for variance, and $d$ is the number of dimensions.

* Theorem 1: $E[||X||^2] = d$.
  * Proof:
  \begin{align}
  E[||X||^2] &= E[X_1^2 + X_2^2 + ... + X_d^2] = E[X_1^2] + E[X_2^2] + ... + E[X_d^2]\\
  &= E[X_1^2 - 0] + E[X_2^2 - 0] + ... + E[X_d^2 - 0] = E[X_1^2 - E[X_1]] + E[X_2^2 - E[X_2]] + ... + E[X_d^2 - E[X_d]]\\
  &= V[X_1] + V[X_2] + ... + V[X_d] = d
  \end{align}
  The last two steps were realized because for a normal standard multivariate distribution, it holds that for each $X_i \sim \mathcal{N}(0,1)$ where $\mathcal{N}(0,1)$ is a standard normal univariate distribution.
* Theorem 2: $E[||X-Y||^2] = 2d$, with  $X,Y \sim \mathcal{N}(0,1,\mathbb{R}^d)$ and &X,Y& independent random variables.
  * Proof:
  \begin{align}
  E[||X-Y||^2] &= E\left[\sum_{i=1}^d (X_i - Y_i)^2\right] = \sum_{i=1}^d E\left[(X_i - Y_i)^2\right]\\
  &= \sum_{i=1}^d E\left[X_i^2 -2X_iY_i + Y_i^2\right] = \sum_{i=1}^d E[X_i^2] -2E[X_i]E[Y_i] + E[Y_i^2]\\
  &= \sum_{i=1}^d E[X_i^2-0] -2(0)(0) + E[Y_i^2 -0] = \sum_{i=1}^d E[X_i^2-E[X_i]] + E[Y_i^2 -E[Y_i]]\\
  &= \sum_{i=1}^d V[X_i] + V[Y_i] = \sum_{i=1}^d 2 = 2d\\
  \end{align}

Finally, we are going to prove that $E[\langle X,Y \rangle] = 0$, with  $X,Y \sim \mathcal{N}(0,1,\mathbb{R}^d)$ and $X,Y$ independent random variables:
The cosine rule stands for:
\begin{align}
  \|X-Y\|^2 = \|X\|^2 + \|Y\|^2 - 2\|X|\|Y\|\cos θ
\end{align}
Considering that $\langle X,Y \rangle = \|X|\|Y\|\cos θ$, we have $\cos θ = \frac{\langle X,Y \rangle}{ \|X|\|Y\|}$, therefore, replacing in the previous equation:
\begin{align}
  \|X-Y\|^2 &= \|X\|^2 + \|Y\|^2 - 2\|X|\|Y\|\left( \frac{\langle X,Y \rangle}{ \|X|\|Y\|} \right)\\
  &= \|X\|^2 + \|Y\|^2 - 2\langle X,Y \rangle\\
\end{align}
Consequently:
\begin{align}
   \langle X,Y \rangle = \frac{1}{2}\left( \|X\|^2 + \|Y\|^2 - \|X-Y\|^2 \right) \\
\end{align}
Thus:
\begin{align}
   E[\langle X,Y \rangle] &= \frac{1}{2}E\left[ \|X\|^2 + \|Y\|^2 - \|X-Y\|^2 \right] \\
   &= \frac{1}{2}\left( E[\|X\|^2] + E[\|Y\|^2] - E[\|X-Y\|^2] \right) = \frac{1}{2}\left( d + d - 2d \right) = 0 \\
\end{align}

##**Problem 8.7**
###**Answer**

The gamma function and its respective aproximation by the Stirling formula are as follows:
\begin{align}
Γ[x]=\int_{0}^{∞} t^{x-1} e^{-t} dt ≈ \sqrt{2π(x-1)}\left(\frac{x-1}{e}\right)^{x-1}
\end{align}
Therefore the volume of a hyperspehere with radius $r$ in $D$ dimensions using the Stirling formula is given by:
\begin{align}
\text{Vol}[r] &≈ \frac{r^D \pi^{D/2}}{\sqrt{2\pi(D/2)}}\left(\frac{D/2}{e}\right)^{D/2}\\
&= \frac{r^D \pi^{D/2}}{\sqrt{\pi D}}\left(\frac{D}{2e}\right)^{D/2}\\
\end{align}
For radius $r=0.5$ we have:
\begin{align}
\text{Vol}[0.5] &≈ \frac{\frac{1}{2}^D \pi^{D/2}}{\sqrt{\pi D}\left(\frac{D}{2e}\right)^{D/2}} = \frac{\frac{\pi}{4}^{D/2}}{\sqrt{\pi D}\left(\frac{D}{2e}\right)^{D/2}}\\
&= \frac{1}{\sqrt{\pi D}\left(\frac{D}{2e}\right)^{D/2}\frac{4}{\pi}^{D/2}} = \frac{1}{\sqrt{\pi D}\left(\frac{2D}{e\pi}\right)^{D/2}}\\
&= \frac{1}{\sqrt{\pi D \left(\frac{2D}{e\pi}\right)^D}}
\end{align}
We can observe that as $D$ increases, the volume aproaches $0$:
\begin{align}
\lim_{D \to ∞}  \text{Vol}[0.5]&= \lim_{D \to ∞}  \frac{1}{\sqrt{\pi D \left(\frac{2D}{e\pi}\right)^D}} = 0
\end{align}

\begin{align}

##**Problem 8.9**
###**Answer**

You can see in "Lecture Notes on High-Dimensional Data" by Sven-Ake Wegner that it is possible to describe a normal distribution for the distance of the samples from the origin. As $D$ increases, the median approaches $\sqrt{D}$ and the variance is always lower or equal to $2$. In these notes, you can find the proof for:

\begin{align}
E[\|X\| - \sqrt{D}] \leq \frac{1}{\sqrt{D}}\\
V[\|X\|] \leq 2\\
\end{align}

We have used the same notation as the answer to Problem 8.6. The next is the code required for the histograms of the samples distance:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the dimensions D
dimensions = [25, 100, 500]

# Number of samples
num_samples = 5000

# Function to calculate Euclidean distance to the center
def calculate_distance_to_center(samples):
    return np.linalg.norm(samples, axis=1)

# Generate samples and make histograms for each dimension
for D in dimensions:
    samples = np.random.multivariate_normal(np.zeros(D), np.eye(D), size=num_samples)
    distances = calculate_distance_to_center(samples)

    plt.figure()
    plt.hist(distances, bins=50, edgecolor='black')
    plt.title(f'Distance Histogram - D={D}')
    plt.xlabel('Distance to center')
    plt.ylabel('Frequency')
    plt.show()
```
