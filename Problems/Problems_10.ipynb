\begin{align}
z_i = \mathbf{ω_1}\cdot (x_{2i-3}, x_{2i-2}, x_{2i-1}, x_{2i}, x_{2i+1})^T

\begin{align}
y_i = \mathbf{ω_2}\cdot (z_{i-1}, z_{i}, z_{i+1})^T
\end{align}
* Let's see that the receptive field of $\mathbf{H}_2$ goes from $z_{i-1}$ to $z_{i+1}$.

  Then, we are going to express $z_{i-1}$ to $z_{i+1}$ in terms of $x_k$ with $k \in \mathbb{Z}$, trying to cover the entire receptive field (the arrow represent the coverage):
  \begin{align}
  z_{i-1} \stackrel{from}{\longrightarrow} x_{2(i-1)-3} = x_{2i-5}\\
