##**Problem 3.3**
Using the non-negative homogeneity property of the ReLU function (see problem 3.5), show that:
\begin{align}
ReLU\left[\beta_1 + \lambda_1 \cdot Ω_1 ReLU\left[\beta_0 + \lambda_0 \cdot Ω_0 x \right]\right]   = \lambda_0 \lambda_1 \cdot ReLU \left[\frac{1}{\lambda_0 \lambda_1}\beta_1 +   Ω_1 \cdot ReLU \left[\frac{1}{\lambda_0}\beta_0 +  Ω_0 x \right]\right]
\end{align}
where $\lambda_0$ and $\lambda_1$ are non-negative scalars. From this, we see that the weight matrices can be
rescaled by any magnitude as long as the biases are also adjusted, and the scale factors can be re-applied at the end of the network
###**Answer**
We can start from the right side expression, using the non-negative property of the ReLU function:
\begin{align}
  \lambda_0 \lambda_1 \cdot ReLU \left[\frac{1}{\lambda_0 \lambda_1}\beta_1 +   Ω_1 \cdot ReLU \left[\frac{1}{\lambda_0}\beta_0 +  Ω_0 x \right]\right]\\
  =  ReLU \left[\frac{\lambda_0 \lambda_1}{\lambda_0 \lambda_1}\beta_1 +   \lambda_0 \lambda_1 \cdot Ω_1 \cdot ReLU \left[\frac{1}{\lambda_0}\beta_0 +  Ω_0 x \right]\right]\\
  = ReLU \left[\beta_1 + \lambda_1 \cdot Ω_1 \cdot ReLU \left[\frac{\lambda_0}{\lambda_0}\beta_0 +  \lambda_0 \cdot Ω_0 x \right]\right]\\
  = ReLU \left[\beta_1 + \lambda_1 \cdot Ω_1 \cdot ReLU \left[\beta_0 +  \lambda_0 \cdot Ω_0 x \right]\right]\\
\end{align}

##**Problem 4.6**
Consider a network with $D_i = 1$ input, $D_o = 1$ output, and $K = 10$ layers, with $D = 10$ hidden units in each. Would the number of weights increase more if we increased the depth by one or the width by one? Provide your reasoning

###**Answer**
Let's see the parameters that this kind of network will have. If K is the number of layers, and D the number of hidden units in each:
\begin{align}
h_1 = \mathrm{a}[\mathrm{β}_0 + Ω_0]; & \quad \mathrm{β}_0 \text{ size is }D*1; & \quad Ω_0 \text{ size is }D*1\\
h_2 = \mathrm{a}[\mathrm{β}_1 + Ω_1 h_1]; & \quad \mathrm{β}_1 \text{ size is }D*1; & \quad Ω_1 \text{ size is }D*D\\
h_3 = \mathrm{a}[\mathrm{β}_2 + Ω_2 h_2]; & \quad \mathrm{β}_2 \text{ size is }D*1; & \quad Ω_2 \text{ size is }D*D\\
\cdot\\
\cdot\\
\cdot\\
h_k = \mathrm{a}[\mathrm{β}_{k-1} + Ω_{k-1} h_{k-1}]; & \quad \mathrm{β}_{k-1} \text{ size is }D*1; & \quad Ω_{k-1} \text{ size is }D*D\\
y = \mathrm{β}_k + Ω_k h_k; & \quad \mathrm{β}_k \text{ size is }1*1; & \quad Ω_k \text{ size is }1*D\\
\end{align}
Thus, the entire network has this amount of parameters $P$:

\begin{align}
P = D+ D + (D^2 + D)(K-1) + 1 + D  = (D^2 + D)(K-1) + 3D + 1
\end{align}

Then if $D=10$ and $K=11$, we obtain $P=1131$. And if $K=10$ and $D=11$, we obtain $P=1222$. Consenquently, the number of parameters increase more if we increase the number of hidden units $D$ than the number of layers $K$. Besides, if we fixed the number of layers $K$, then the number of parameters will be represented by a quadratic function of $D$, and if we fixed the number of hidden units $D$, then we will obtain a linear function of $K$ for representing the number of parameters. Let's see a graph when we fixed one of $D=10$ or $K=10$:


###Graph

```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos los valores de D y K
D_values = np.linspace(1, 15, 100)  # Rango de valores positivos para D
K = 10  # Valor fijo positivo de K

# Función a graficar
def func(D, K):
    return (D**2 + D)*(K-1) + 1 + 3*D

# Graficamos la función fijando K
plt.figure(figsize=(10, 6))
plt.plot(D_values, func(D_values, K), label=r'$K$ fixed = {}'.format(K))

# Graficamos la función fijando D
D = 10  # Valor fijo positivo de D
K_values = np.linspace(1, 15, 100)  # Rango de valores positivos para K
plt.plot(K_values, func(D, K_values), label=r'$D$ fixed = {}'.format(D))
plt.xlabel('$D$ with $K$ fixed or $K$ with $D$ fixed')
plt.ylabel(r'Number of parameters')
plt.title(r'Graph of number of parameters with $K$ or $D$ fixed')
plt.legend()

plt.show()
```

##**Problem 4.8**
Figure 4.9 shows the activations in the three hidden units of a shallow network
(as in figure 3.3). The slopes in the hidden units are 1.0, 1.0, and -1.0, respectively, and the “joints” in the hidden units are at positions 1/6, 2/6, and 4/6. Find values of $\phi_0, \phi_1, \phi_2$, and $\phi_3$ that will combine the hidden unit activations as $\phi_0 + \phi_1 h_1 + \phi_2 h_2 + \phi_3 h_3$ to create a function with four linear regions that oscillate between output values of zero and one. The slope of the leftmost region should be positive, the next one negative, and so on. How many linear regions will we create if we compose this network with itself? How many will we create if we compose it with itself K times?

###**Answer 1**
From the problem 3.3 we obtained these expressions for the position of the joints:
\begin{align}
  x_1 = -\frac{\phi_{10}}{\phi_{11}}\\
  x_2 = -\frac{\phi_{20}}{\phi_{21}}\\
  x_3 = -\frac{\phi_{30}}{\phi_{31}}\\
\end{align}
Now, knowing the slopes of the hidden units, we can determine the values for the biases of each hidden unit. We have $\theta_{11} = \theta_{21} = 1$ and $\theta_{31} = -1$. Then using the equations above we obtain for each hidden unit:
\begin{align}
  1/6 = -\frac{\phi_{10}}{\phi_{11}} = -\frac{\phi_{10}}{1}; \quad & \phi_{10}=-\frac{1}{6}\\
  2/6 = -\frac{\phi_{20}}{\phi_{21}} = -\frac{\phi_{20}}{1}; \quad & \phi_{20}=-\frac{2}{6}\\
  4/6 = -\frac{\phi_{30}}{\phi_{31}} = -\frac{\phi_{30}}{-1}; \quad & \phi_{30}=\frac{4}{6}
\end{align}
From the figure 4.9 we can determine which hidden units will be activated at the output of our shallow neural network:
- 1° Linear region: $h_3$ active.
- 2° Linear region: $h_1$ and $h_3$ active.
- 3° Linear region: $h_1, h_2$ and $h_3$ active.
- 4° Linear region: $h_1$ and $h_2$ active.

Aditionally, we are required to obtain an output where these linear regions oscillate between output values of zero and one, starting with the slope of the first linear region negative, second linear region positive and so on, so we have:

- Slope 1° Linear region: $S_{r_1} = \frac{1-0}{1/6-0}=\frac{1}{1/6}=6$
- Slope 2° Linear region: $S_{r_2} = \frac{0-1}{2/6-1/6}=\frac{-1}{1/6}=-6$
- Slope 3° Linear region: $S_{r_3} = \frac{1-0}{4/6-2/6}= \frac{1}{2/6} = 3$
- Slope 4° Linear region: $S_{r_4} = \frac{0-1}{1-4/6}=\frac{-1}{2/6}=-3$

Finally with this information we can determine the values of the weights for each hidden unit:
\begin{align}
  S_{r_1} = \phi_3 \theta_{31}; &\quad \phi_3 = \frac{S_{r_1}}{\theta_{31}} = \frac{6}{-1}=-6\\
  S_{r_2} = \phi_1 \theta_{11} + \phi_3 \theta_{31}; &\quad \phi_1 = \frac{S_{r_2} - \phi_3 \theta_{31}}{\theta_{11}} = \frac{-6 - 6}{1}=-12\\
  S_{r_3} = \phi_1 \theta_{11} + \phi_2 \theta_{21} + \phi_3 \theta_{31}; &\quad \phi_2 = \frac{S_{r_3} - \phi_1 \theta_{11} - \phi_3 \theta_{31}}{\theta_{21}} = \frac{3 + 12 - 6}{1}=9\\
\end{align}

For the last parameter $\phi_0$ we just need to set the equation of the first linear region starting at $x=0, y=0$. The first linear region equation can be obtained with the preactivated function of the third hidden unit multiplied by its weight plus $\theta_0$: $y= \phi_3(4/6+\theta_{31}x) + \theta_0=-6(4/6 - x)=-4-6x + \theta_0$. With $x=0$ then $y= -4 + \theta_0$. Then with $\theta_0  = 4$ we obtain $y_{r_1}=0$.

```python
# Imports math library
import numpy as np
# Imports plotting library
import matplotlib.pyplot as plt
```

```python
# Define the Rectified Linear Unit (ReLU) function
def ReLU(preactivation):
  activation = np.zeros_like(preactivation);
  activation = np.clip(preactivation,0, None)
  return activation
```

```python
# Define a shallow neural network with, one input, one output, and three hidden units
def shallow_1_1_3(x, activation_fn, phi_0,phi_1,phi_2,phi_3, theta_10, theta_11, theta_20, theta_21, theta_30, theta_31):
  pre_1 = theta_10 + theta_11 * x
  pre_2 = theta_20 + theta_21 * x
  pre_3 = theta_30 + theta_31 * x

  act_1 = activation_fn(pre_1)
  act_2 = activation_fn(pre_2)
  act_3 = activation_fn(pre_3)

  w_act_1 = phi_1 * act_1
  w_act_2 = phi_2 * act_2
  w_act_3 = phi_3 * act_3

  y = w_act_1 + w_act_2 + w_act_3 + phi_0

  return y, pre_1, pre_2, pre_3, act_1, act_2, act_3, w_act_1, w_act_2, w_act_3
```

```python
# Plot the shallow neural network.  We'll assume input in is range [0,1] and output [-1,1]
# If the plot_all flag is set to true, then we'll plot all the intermediate stages as in Figure 3.3
def plot_neural(x, y, pre_1, pre_2, pre_3, act_1, act_2, act_3, w_act_1, w_act_2, w_act_3, plot_all=False, x_data=None, y_data=None):

  # Plot intermediate plots if flag set
  if plot_all:
    fig, ax = plt.subplots(3,3)
    fig.set_size_inches(8.5, 8.5)
    fig.tight_layout(pad=3.0)
    ax[0,0].plot(x,pre_1,'r-'); ax[0,0].set_ylabel('Preactivation')
    ax[0,1].plot(x,pre_2,'b-'); ax[0,1].set_ylabel('Preactivation')
    ax[0,2].plot(x,pre_3,'g-'); ax[0,2].set_ylabel('Preactivation')
    ax[1,0].plot(x,act_1,'r-'); ax[1,0].set_ylabel('Activation')
    ax[1,1].plot(x,act_2,'b-'); ax[1,1].set_ylabel('Activation')
    ax[1,2].plot(x,act_3,'g-'); ax[1,2].set_ylabel('Activation')
    ax[2,0].plot(x,w_act_1,'r-'); ax[2,0].set_ylabel('Weighted Act')
    ax[2,1].plot(x,w_act_2,'b-'); ax[2,1].set_ylabel('Weighted Act')
    ax[2,2].plot(x,w_act_3,'g-'); ax[2,2].set_ylabel('Weighted Act')

    for plot_y in range(3):
      for plot_x in range(3):
        ax[plot_y,plot_x].set_xlim([0,1]);ax[plot_x,plot_y].set_ylim([-1,1])
        ax[plot_y,plot_x].set_aspect(0.5)
      ax[2,plot_y].set_xlabel('Input, $x$');
    plt.show()

  fig, ax = plt.subplots()
  ax.plot(x,y)
  ax.set_xlabel('Input, $x$'); ax.set_ylabel('Output, $y$')
  ax.set_xlim([0,1]);ax.set_ylim([-1,1])
  ax.set_aspect(0.5)
  if x_data is not None:
    ax.plot(x_data, y_data, 'mo')
    for i in range(len(x_data)):
      ax.plot(x_data[i], y_data[i],)
  plt.show()
```

```python
# Now lets define some parameters and run the neural network
theta_10 =  -1/6 ; theta_11 = 1
theta_20 = -2/6 ; theta_21 = 1
theta_30 = 4/6  ; theta_31 = -1
phi_0 = 4; phi_1 = -12; phi_2 = 9; phi_3 = -6

# Define a range of input values
x = np.arange(0,1,0.01)

# We run the neural network for each of these input values
y, pre_1, pre_2, pre_3, act_1, act_2, act_3, w_act_1, w_act_2, w_act_3 = \
    shallow_1_1_3(x, ReLU, phi_0,phi_1,phi_2,phi_3, theta_10, theta_11, theta_20, theta_21, theta_30, theta_31)
# And then plot it
plot_neural(x, y, pre_1, pre_2, pre_3, act_1, act_2, act_3, w_act_1, w_act_2, w_act_3, plot_all=True)
```

There 3 regions with slopes required: $S_{r_1}>0, S_{r_2}< 0$ and $S_{r_3}>0$ However, there are only four possible cases for activation of the hidden units, as we need to have 3 joints:
* First case:

  * for $h_2$: inactivated on $r_1$, inactivated on $r_2$, activated on $r_3$

##**Problem 4.9**
Consider two neural networks that map a scalar input $x$ to a scalar output $y$.
The first network is shallow and has $D = 95$ hidden units. The second is deep and has $K = 10$ layers, each containing $D = 5$ hidden units. How many parameters does each network have? How many linear regions can each network make? Which would run faster?
###**Answer**
For a shallow neural network with $D$ hidden units, one scalar input value and one scalar output value, each hidden unit contributes 3 parameter, input weight, bias and output weight, totaling $3*D$ paramaters. Adding the output bias gives a total of $3*D + 1$ parameters. Thus, our shallow neural network has $286$ parameters. Similarly, using the equation of problem 4.9 above our deep neural network has also $286$ paramteres.

The shallow neural network with one scalar input and scalar output value can make $D+1$ linear regions, resulting for our shallow neural network 96 linear regions. The deep neural network with one scalar input and one scalar output value can make $(D+1)^{K}$ linear regions. Therefore, our deep neural network can make $6^{10}=60466176$ linear regions.

Shallow neural network run faster.
