##**Problem 9.1**
###**Answer**

\begin{align}
\hat{\mathbf{ϕ}} &= \underset{\mathbf{ϕ}}{\text{argmax}}\left[ \prod_{i=1}^I Pr(y_i|x_i,\mathbf{ϕ}) Pr(\mathbf{ϕ}) \right] = \underset{\mathrm{\mathbf{ϕ}}}{\text{argmin}}\left[- \sum_{i=1}^I \log \left(Pr(y_i|x_i,\mathbf{ϕ}) Pr(\mathbf{ϕ})\right) \right]\\
&= \underset{\mathrm{\mathbf{ϕ}}}{\text{argmin}}\left[- \sum_{i=1}^I \log \left(Pr(y_i|x_i,\mathbf{ϕ})\right) - \sum_{i=1}^I \log \left(Pr(\mathbf{ϕ})\right) \right] = \underset{\mathrm{\mathbf{ϕ}}}{\text{argmin}}\left[L[\mathbf{ϕ}] - I \log \left(Pr(\mathbf{ϕ})\right) \right]\\
&= \underset{\mathrm{\mathbf{ϕ}}}{\text{argmin}}\left[L[\mathbf{ϕ}] - I \log \left(\prod_{j=1}^J \text{Norm}_{ϕ_j}[0,σ^2_\mathbf{ϕ}]\right) \right] = \underset{\mathbf{ϕ}}{\text{argmin}}\left[L[\mathbf{ϕ}] - I \log \left(\prod_{j=1}^J \frac{1}{\sqrt{2\pi σ^2_\mathbf{ϕ}}}\exp\left[-\frac{(ϕ_j - μ)^2}{2σ^2_\mathbf{ϕ}}\right] \right) \right]\\
&= \underset{\mathbf{ϕ}}{\text{argmin}}\left[L[\mathbf{ϕ}] - I \sum_{j=1}^J \log \left(\frac{1}{\sqrt{2\pi σ^2_\mathbf{ϕ}}}\right) - I  \sum_{j=1}^J \log \left(\exp\left[-\frac{(ϕ_j - μ)^2}{2σ^2_\mathbf{ϕ}}\right] \right) \right]\\
&= \underset{\mathbf{ϕ}}{\text{argmin}}\left[L[\mathbf{ϕ}] + I J  \log \left(\sqrt{2\pi σ^2_\mathbf{ϕ}}\right) + \frac{I}{2σ^2_\mathbf{ϕ}}  \sum_{j=1}^J ϕ_j^2\right]\\
\end{align}
If we assume that $σ^2_\mathbf{ϕ},\ I$ and $J$ are fixed, with $\lambda = \frac{I}{2σ^2_\mathbf{ϕ}}$ then:
\begin{align}
\hat{\mathbf{ϕ}} &=  \underset{\mathbf{ϕ}}{\text{argmin}}\left[L[\mathbf{ϕ}] + \lambda  \sum_{j=1}^J ϕ_j^2\right]\\
\end{align}

##**Problem 9.2**
###**Answer**

We have that $L2$ reguralization is given by:
\begin{align}
L_2[\mathbf{ϕ}] = L[\mathbf{ϕ}] + \lambda \sum_{j=1}^J ϕ_j^2
\end{align}
where $L[\mathbf{ϕ}] = \sum_{i=1}^I \ell_i [\mathbf{x}_i,\mathbf{y}_i]$. The gradient of $L2$ with respect to $\mathbf{ϕ}$ is given by:
\begin{align}
\frac{\partial L_2[\mathbf{ϕ}]}{\partial \mathbf{ϕ}} = \frac{L[\mathbf{ϕ}]}{\partial \mathbf{ϕ}} + \lambda \sum_{j=1}^J \frac{ϕ_j^2 }{\partial \mathbf{ϕ}}= \frac{L[\mathbf{ϕ}]}{\partial \mathbf{ϕ}} + 2\lambda\mathbf{ϕ}
\end{align}
The last step was done because $\sum_{j=1}^J \frac{ϕ_j^2 }{\partial \mathbf{ϕ}}$ is a sum of matrices with same shape as $\mathbf{ϕ}$, where for each $j$ all the values of the matrix are $0$, except at the $j$-th place where the value is $2ϕ_j$.
Therefore, the change of the gradient is an increment given by $2\lambda\mathbf{ϕ}$.

##**Problem 9.3**
###**Answer**
This loss function with added noise to the input x is given by:
\begin{align}
L'[\mathbf{ϕ}] = \sum_{i=1}^I (\theta_0 + \theta_1 (x_i + ϵ_i) - y_i)^2
\end{align}
where $ϵ_i$ represents the noise added to each x_i and $ϵ_i \sim \mathcal{N}[0,\sigma_x^2]$, therefore:
\begin{align}
L'[\mathbf{ϕ}] &= \sum_{i=1}^I (\theta_0 + \theta_1 x_i - y_i + \theta_1 ϵ_i)^2\\
&= \sum_{i=1}^I (\theta_0 + \theta_1 x_i - y_i)^2 + \sum_{i=1}^I 2(\theta_0 + \theta_1 x_i - y_i)\theta_1ϵ_i + \sum_{i=1}^I (\theta_1ϵ_i)^2\\
&= L[\mathbf{ϕ}] + 2\theta_1\sum_{i=1}^I (\theta_0 + \theta_1 x_i - y_i)ϵ_i + \theta_1^2\sum_{i=1}^I ϵ_i^2\\
\end{align}
If we calculate the gradient, we obtain:
\begin{align}
\frac{\partial L'[\mathbf{ϕ}]}{\partial \theta_0} &= \frac{\partial L[\mathbf{ϕ}]}{\partial \theta_0} + 2\theta_1\sum_{i=1}^I ϵ_i \\
\frac{\partial L'[\mathbf{ϕ}]}{\partial \theta_1} &= \frac{\partial L[\mathbf{ϕ}]}{\partial \theta_1} + 2\theta_1\sum_{i=1}^I x_i ϵ_i + 2\theta_1\sum_{i=1}^I ϵ_i^2\\
\end{align}
Now if we take expectations, and knowing that $E[ϵ_i] = 0$, $E[(ϵ_i-E[ϵ_i])^2] = E[ϵ_i^2]-E[ϵ_i]^2 = E[ϵ_i^2] = \sigma_x^2$, we have:
\begin{align}
E\left[\frac{\partial L'[\mathbf{ϕ}]}{\partial \theta_0}\right] &= E\left[\frac{\partial L[\mathbf{ϕ}]}{\partial \theta_0}\right] + E\left[ 2\theta_1\sum_{i=1}^I ϵ_i \right] \\
&= \frac{\partial L[\mathbf{ϕ}]}{\partial \theta_0} + 2\theta_1\sum_{i=1}^I E\left[ϵ_i \right] = \frac{\partial L[\mathbf{ϕ}]}{\partial \theta_0}\\
E\left[\frac{\partial L'[\mathbf{ϕ}]}{\partial \theta_1}\right] &= E\left[\frac{\partial L[\mathbf{ϕ}]}{\partial \theta_1}\right] + E\left[2\theta_1\sum_{i=1}^I x_i ϵ_i\right] + E\left[2\theta_1\sum_{i=1}^I ϵ_i^2\right]\\
&= \frac{\partial L[\mathbf{ϕ}]}{\partial \theta_1} + 2\theta_1\sum_{i=1}^I x_i E\left[ϵ_i\right] + 2\theta_1\sum_{i=1}^I E\left[ϵ_i^2\right]\\
&= \frac{\partial L[\mathbf{ϕ}]}{\partial \theta_1}  + 2\theta_1\sum_{i=1}^I \sigma_x^2 = \frac{\partial L[\mathbf{ϕ}]}{\partial \theta_1}  + 2\theta_1 I \sigma_x^2\\
\end{align}
Therefore, the gradient update is just an increment of the value of partial derivative of the loss with respect to the parameter $\theta_1$ by $2\theta_1 I \sigma_x^2$, but not the loss with respect to the parameter $\theta_0$. However, $L2$ regularization, also referred to as "weight decay" sometimes, only takes into account the weights and not the bias. Therefore, in this case, this gradient update is equivalent to the gradient update in $L2$ regularization.

The following is the standard gradient update using $L2$ regularization:
\begin{align}
\mathbf{ϕ} &← \mathbf{ϕ} - α\left( \frac{\partial \tilde{L}}{\partial \mathbf{ϕ}} \right) = \mathbf{ϕ} - α\left( \frac{\partial L}{\partial \mathbf{ϕ}} + \frac{λ}{2\alpha} 2 \mathbf{ϕ} \right)\\
